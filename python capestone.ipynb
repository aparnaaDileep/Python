 Discover which team and position have the highest salary expenditure. 

```python
team_salary_expenditure = data.groupby("Team")["Salary"].sum()
position_salary_expenditure = data.groupby("Position")["Salary"].sum()
```

```python
highest_team = team_salary_expenditure.idxmax()
highest_position = position_salary_expenditure.idxmax()
```

```python
print("The team that has highest salary expenditure is ",highest_team)
print("The position that has highest salary expenditure is ", highest_position)
```

```python
plt.figure(figsize=(8, 6))
team_salary_expenditure.sort_values().plot(kind='bar', color="skyblue", edgecolor="black")
plt.title("Total Salary Expenditure by Team", fontsize=14)
plt.xlabel("Total Salary ", fontsize=12)
plt.ylabel("Team", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
position_salary_expenditure.sort_values().plot(kind='bar', color="lightgreen", edgecolor="black")
plt.title("Total Salary Expenditure by Position", fontsize=14)
plt.xlabel("Total Salary (in $)", fontsize=12)
plt.ylabel("Position", fontsize=12)
plt.show()
```
